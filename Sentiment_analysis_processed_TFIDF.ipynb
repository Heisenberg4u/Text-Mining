```python
import nltk                # Natural Language ToolKit, for stopping, stemming, lemmatization
nltk.download('punkt')     # default download
nltk.download('stopwords') # downloading the stopwords to be used in the stopping process
nltk.download('wordnet')   # downloading wordnet(is like imagenet for text) text data for stemming and lemmatization purpose
from nltk.corpus import stopwords  # imports the stopwords we can use
from nltk.stem.wordnet import WordNetLemmatizer   # Lemmatizer
import numpy as np
import re          # Regular Expression, RegEx
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# [^a-z ] pattern .. that all alphabets a-z, a to z and space, caret operator (^) means not
# that if you find any character in the text(3rd value), which is NOT (a-z, space), then replace it by 2nd value

re.sub('[^a-z ]', '', 'abc56knLng$')
```

```python

text = '' # text is an empty string

for i in raw:
    text += re.sub('[^a-z ]', '', i.strip()[2:].lower())+' '
    # text = text + re.sub('[^a-z ]', '', i.strip()[2:].lower())+' '

# i.strip() removes extra spaces from the start and end of a line of text
#   starting from   here and ending here  
# i.strip()[2:] # taking the text part of each line
# i.strip()[2:].lower() # Any capital letter will become smalls by using .lower()
# re.sub --> sub means substitute


print(text)
```

```python
nltk.word_tokenize(text) # returns individual words from the given text in the form of a list, this is tokenizer
```

```python
temp = [1, 2, 3, 3, 4, 2, 4]
list( set(temp) )
```

```python
word = list( set( nltk.word_tokenize(text) ) ) # word is now the list of all the unique words in the 4000 lines of text
print(len(word))
```

```python
word                    # list of all the unique words
```

```python
filtered = list(set([w for w in word if not w in stopwords.words('english')]))
print(len(filtered))
print(filtered)
```

```python
stopwords.words('english')
```

```python
temp = [] # empty list
for w in word: # word is a list of unique words
    if not w in stopwords.words('english'):
        temp.append(w)

filtered = list(set(temp)) # filtered will be the list of all unique non-stop words
print(len(filtered))
print(filtered)
# 1625 - 1515 words were stopwords which have been removed
```

```python
wordnet_lemmatizer = WordNetLemmatizer()

wordnet_lemmatized_v = []

for item in filtered:
    wordnet_lemmatized_v.append(wordnet_lemmatizer.lemmatize(item, 'v'))

wordnet_v = list(set(wordnet_lemmatized_v))
print(len(wordnet_v))
```

```python
train_file_wordnet_v = []
Y_train_raw = []

for line in raw:
    Y_train_raw.append(int(line[0]))
    x = re.sub('[^a-z ]', '', line.strip()[2:].lower())
    tokens = nltk.word_tokenize(x)
    filtered_tokens = [w for w in tokens if not w in stopwords.words('english')]
    stemmed_tokens = ''
    for item in filtered_tokens:
        stemmed_tokens += wordnet_lemmatizer.lemmatize(item, 'v')+' '
    train_file_wordnet_v.append(stemmed_tokens[:-1])

print(len(train_file_wordnet_v))
print(train_file_wordnet_v[1])
```

```python
vectorizer = TfidfVectorizer()
```

```python
X_train_tfidf = vectorizer.fit_transform(train_file_wordnet_v)
```

```python
print(vectorizer.get_feature_names())
print(X_train_tfidf.shape)
print(type(X_train_tfidf))
X_train = X_train_tfidf.toarray()
Y_train = np.array(Y_train_raw)
print(type(X_train))
print(X_train.shape)
```

```python
test_file_wordnet_v = []
Y_test_raw = []

f = open('te')
read_test = f.read().splitlines()
f.close()

for line in read_test:
    Y_test_raw.append(int(line[0]))
    x = re.sub('[^a-z ]', '', line.strip()[2:].lower())
    tokens = nltk.word_tokenize(x)
    filtered_tokens = [w for w in tokens if not w in stopwords.words('english')]
    stemmed_tokens = ''
    for item in filtered_tokens:
        stemmed_tokens += wordnet_lemmatizer.lemmatize(item, 'v')+' '
    test_file_wordnet_v.append(stemmed_tokens[:-1])

print(len(test_file_wordnet_v))
print(test_file_wordnet_v[4])
```

```python
X_test_tfidf = vectorizer.transform(test_file_wordnet_v)
print(X_test_tfidf.shape)
print(type(X_test_tfidf))
X_test = X_test_tfidf.toarray()
Y_test = np.array(Y_test_raw)
print(type(X_test))
print(X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
model_parameters = {
        'n_estimators':[50, 100, 150, 200],
        'criterion':['gini', 'entropy'],
        'max_depth': [3, 5, 7, 9, 11]
    }
```

```python
model = RandomForestClassifier(random_state=1)
gscv = GridSearchCV(estimator=model, 
                    param_grid=model_parameters, 
                    cv=5, 
                    verbose=1, 
                    n_jobs=-1,
                    scoring='roc_auc')

gscv.fit(X_train, Y_train)
```

```python
print('The best parameter are -', gscv.best_params_)
```

```python
print(gscv.best_score_)
print(gscv.best_estimator_)
print(gscv.scorer_)
```

```python
print('AUC on test by gscv =', roc_auc_score(y_true=Y_test,
                                                        y_score=gscv.predict_proba(X_test)[:, 1]))
```
