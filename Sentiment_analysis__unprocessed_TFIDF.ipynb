```python
X_train_raw = [] # Empty list to store the text part of a line
Y_train_raw = [] # Empty list to store the label of a line

# import random
# random.shuffle(raw)

for line in raw:
    X_train_raw.append(line[2:])
    Y_train_raw.append(int(line[0]))

X_train_raw
```

```python
Y_train = np.array(Y_train_raw) # to convert list to numpy array, because sklearn models take t_train as numpy/pandas array only
print(len(X_train_raw)) # X_train_raw is a list, 4000
print(X_train_raw)
print(len(Y_train))
print(Y_train)
print(Y_train.sum()) # out of 4000 labels, 2000 are 0's and 2000 are 1's, so the sum will be 2000
```

```python
# model = RandomForestClassifier()
vectorizer = TfidfVectorizer() # create vectorizer object just like we created model object, model = RandomForestClassifier()
# vectorizer can take input as a list of sentences
# X_train_tfidf = vectorizer.fit_transform(X_train_raw)
vectorizer.fit(X_train_raw) # creates the structure of the tf-idf table, number of rows = number of sentences, number of columns = number of unique words in all the 4000 sentences
X_train_tfidf = vectorizer.transform(X_train_raw) # fill the values in tf-idf tabled formed above


# X_train_tfidf = vectorizer.fit_transform(X_train_raw) is same as 
# 1. vectorizer.fit(X_train_raw) # find the total number of unique words(number of columns in tf-idf table) from the 4000 sentences & also compute idf of each such unique word
# X_train_tfidf = vectorizer.transform(X_train_raw) # compute the tf of words occuring in a sentence and update the tf-idf table
X_train_tfidf # tf-idf table for training text contained in X_train_raw
# X_train_tfidf is not in pandas/numpy format .. it is in a specific format called Compressed Sparse Row format
# but we can get numpy table from this by using X_train_tfidf.toarray() --> become numpy format
```

```python
print(vectorizer.get_feature_names()) # print the unique words found in all the 4000 lines
print(X_train_tfidf.shape) # size of tf-idf table built from training text
# print(type(X_train_tfidf))
X_train = X_train_tfidf.toarray() # converting X_train_tfidf from a different format to numpy format so that we can use it afterwards with any model building exercise
print(type(X_train)) # shows numpy
print(X_train.shape)
# X_train as a numpy tf-idf table and Y_train a numpy array of labels is ready for classification model training
```

```python
f = open('te')
read_test = f.read().splitlines()
f.close()

X_test_raw = []
Y_test_raw = []

# random.shuffle(read_test) # shuffle the lines, and has no impact on classification performance

for line in read_test:
    X_test_raw.append(line[2:])
    Y_test_raw.append(int(line[0]))

Y_test = np.array(Y_test_raw)
print(len(X_test_raw))
print(X_test_raw)
print(len(Y_test))
print(Y_test)
```

```python

X_test_tfidf = vectorizer.transform(X_test_raw) # we are using vectorizer.transform() only and not .fit(), because calculations
# were already done during the .fit(X_train)
print(X_test_tfidf.shape)
print(type(X_test_tfidf))
X_test = X_test_tfidf.toarray() # converts to numpy format
print(type(X_test))
print(X_test.shape)
# X_test and Y_test ready for evaluating classification performance on test data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
model_parameters = {
        'n_estimators':[50, 100, 150, 200],
        'criterion':['gini', 'entropy'],
        'max_depth': [3, 5, 7, 9, 11]
    }
```

```python
model = RandomForestClassifier(random_state=1)
gscv = GridSearchCV(estimator=model, 
                    param_grid=model_parameters, 
                    cv=5, 
                    verbose=1, 
                    n_jobs=-1,
                    scoring='roc_auc')

gscv.fit(X_train, Y_train)
```

```python
print('The best parameter are -', gscv.best_params_)
```

```python
print(gscv.best_score_)
print(gscv.best_estimator_)
print(gscv.scorer_)
```

```python
print('AUC on test by gscv =', roc_auc_score(y_true=Y_test,
                                                        y_score=gscv.predict_proba(X_test)[:, 1]))
```
